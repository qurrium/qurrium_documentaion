# 3.2 Magnetization Square for Any Directions

## Basic Usage


### a. Import the instances


```python
from qurry import MagnetSquare

experiment_magnet_square = MagnetSquare()
```

### b. Preparing quantum circuit


```python
from qurry.recipe import trivial_paramagnet, ghz
```

```python
circuits_dict = {
    "trivialPM_2": trivial_paramagnet(2),
    "trivialPM_4": trivial_paramagnet(4),
    "trivialPM_6": trivial_paramagnet(6),
    "trivialPM_8": trivial_paramagnet(8),
    "cat_2": ghz(2),
    "cat_4": ghz(4),
    "cat_6": ghz(6),
    "cat_8": ghz(8),
}
print("| trivial paramagnet and cat in 4 qubits:")
print(circuits_dict["trivialPM_4"])
print(circuits_dict["cat_4"])
```

### c. Execute the circuit


#### i. Directly input the circuit

After executing, it will return a uuid of experiment. You can use this uuid to get the result of the experiment.


```python
exp1 = experiment_magnet_square.measure(
    circuits_dict["cat_6"], unitary_operator="z", shots=4096
)
exp1
```

Each experiment result will be stored in a container `.exps`, which you can access later by the uuid of the experiment. For example, you can access the result of `exp1` by `experiment_magnet_square.exps[exp1.exp_id]`.


```python
print("| The exp_id:", exp1.exp_id)
experiment_magnet_square.exps[exp1.exp_id]
```

For `MagnetSquare`, its `.analyze` does not require any arguments, so its post-processing will be executed automatically after `.measure`.


```python
exp1.reports
```

```python
report01 = exp1.reports[0]
report01
```

```python
report01.results_fields()
```

```python
print("| The purity:")
print(report01["default", "magnet_square"])
# or
print(report01.results["default"].magnet_square)
```

#### ii. Add the circuits to container `.waves`, then call them later.

Since we have executed an experiment, the circuit we input in `exp1` is stored in the container `.waves` with serial number `0`.


```python
experiment_magnet_square.waves
```

But we can also add the circuit to the container `.waves` with a custom name.
The name should be unique, otherwise it will be overwritten.
The method `add` will return the actual name of the circuit in the container.


```python
for k, v in circuits_dict.items():
    key_of_cirq = experiment_magnet_square.add(v, k)
    print(f"| {k} added with key: {key_of_cirq}")
print(experiment_magnet_square.waves["cat_4"])
```

If there is a circuit with the same name, it will be replaced by the new one.


```python
print(experiment_magnet_square.add(circuits_dict["cat_4"], "cat_4"))
print(experiment_magnet_square.waves["cat_4"])
```

Otherwise, you will need to use `replace="duplicate"` to prevent it from being replaced.


```python
duplicated_case01 = experiment_magnet_square.add(
    circuits_dict["cat_4"], "cat_4", replace="duplicate"
)
print(duplicated_case01)
print(experiment_magnet_square.waves[duplicated_case01])
```

Now we have prepared the circuit and stored it in the container `.waves`.


```python
experiment_magnet_square.waves
```

Finally, we can execute the circuit and get the result.


```python
exp2 = experiment_magnet_square.measure("cat_4", unitary_operator="z", shots=4096)
exp2
```

```python
report02 = exp2.reports[0]
report02
```

### d. More direction Consider

`MagnetSquare` also can measure on different direction. Except `x`, `y`, `z` have been preset. You can use other single qubit instruction like `Gate` from `qiskit.circuit` or `Operator` from `qiskit.quantum_info` to set specific direction you wanted.

For example, we use a `RZGate` on **$\phi$ = 0** which is identity and its `Operator` version.


```python
from qiskit.circuit.library import RZGate
from qiskit.quantum_info import Operator
```

Preparing `RZGate` and its `Operator` version.


```python
rz_0 = RZGate(0)
rz_0_matrix = rz_0.to_matrix()
rz_0_matrix
```

```python
rz_0_op = Operator(rz_0_matrix)
rz_0_op
```

Executing the experiments


```python
exp_rz_gate = experiment_magnet_square.measure(
    "cat_4", unitary_operator=rz_0, shots=4096
)
reports_rz_gate = exp_rz_gate.reports[0]
print("| magnetization square using RZGate")
print("|" + "-" * 50)
print("| magnetization square", reports_rz_gate["default", "magnet_square"])
# or
print("| magnetization square", reports_rz_gate.results["default"].magnet_square)
```

```python
exp_rz_op = experiment_magnet_square.measure(
    "cat_4", unitary_operator=rz_0_op, shots=4096
)
reports_rz_op = exp_rz_op.reports[0]
print("| magnetization square using RZGate Operator")
print("|" + "-" * 50)
print("| magnetization square", reports_rz_op["default", "magnet_square"])
# or
print("| magnetization square", reports_rz_op.results["default"].magnet_square)
```
